```python
#According to 2022 -->Most wicket takers
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Most Wickets in Test Cricket .csv")
df
```

```python
import plotly.express as px
```

```python
df.columns
```

```python
px.pie(df,names="Player ",values="Wickets ",hole=0.4)
```

```python
px.sunburst(df,path=["Average","Player ","Wickets "],values="Matches")
```

```python
# Assuming 'df' is your DataFrame with columns 'Player' and 'Wickets'
top_5_players = df.sort_values(by="Wickets ", ascending=False).head(5)
fig = px.bar(top_5_players,x='Player ',y="Wickets ",title='Top 5 Players with Most Wickets')
fig.show()
```

```python
fig=px.pie(df,names="Player ",values="Wickets ")
fig.update_traces(textposition="inside")
fig.show()
```
